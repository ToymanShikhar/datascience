# Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)
```

```python
y1 = x**2
y2 = 2*x + 3
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x,y1,color='red',label='Apple',linestyle='dashed',marker='o')
plt.plot(x,y2,color='green',label='Kiwi', marker='*')
plt.xlabel("Time")
plt.ylabel('Price')
plt.title("Fruits Price")
plt.legend()
plt.show()
```

# Scatter Plots

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y1,color='red',label='Apple',linestyle='dashed',marker='o')
plt.scatter(x,y2,color='green',label='Kiwi', marker='*')
plt.xlabel("Time")
plt.ylabel('Price')
plt.title("Fruits Price")
plt.legend()
plt.show()
```

# Bar Plots

```python
plt.style.use('dark_background')
x_coordinates = np.array([0,2,4])
plt.bar(x_coordinates,np.array([10,7,5]),label='Current Year',tick_label=["Gold",'Silver','Platinum'])
plt.bar(x_coordinates+0.8,np.array([4,6,14]),label='Next Year')
plt.ylim(0,20)
plt.legend()
plt.show()
```

# Pie Charts

```python
plt.figure(figsize=(8,8))
plt.style.use("seaborn")
subjects = 'Physics','Chemistry','Maths','English'
marks = [20,34,23,15]
plt.pie(marks,labels=subjects,explode=(0,0,0.1,0),shadow=True)
plt.title('Marks')
plt.show()
```

# Histograms

```python
Xsn = np.random.randn(100)
sigma = 8
mean = 70
X = np.round(Xsn*sigma + mean)
print(X)
```

```python
plt.style.use("seaborn")
plt.hist(X)
plt.ylabel("Frequency")
plt.xlabel("Marks")
plt.title("Physics Marks")
plt.show()
```
