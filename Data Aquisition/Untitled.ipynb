## Image Scraping with bs4

```python
!pip3 install html5lib
```

```python
from bs4 import BeautifulSoup as soup
```

```python
url_img = "https://www.passiton.com/inspirational-quotes"
```

```python
response = requests.get(url_img)
```

```python
print(response)
```

```python
print(response.content)
```

```python
data = soup(response.content,'html.parser')
```

```python
print(data)
```

```python
soup.findAll?
```

```python
images = data.findAll('div',attrs={'class':'col-6 col-lg-3 text-center margin-30px-bottom sm-margin-30px-top'})
```

```python
images
```

```python
images[0]
```

```python
a = requests.get(images[0].img.attrs['src'])
```

```python
a.content
```

```python
for idx,imgs in enumerate(images):
    with open('inspiration{}.jpg'.format(idx), 'wb') as file:
        url = imgs.img.attrs['src']
        df = requests.get(url)
        file.write(df.content)
```
