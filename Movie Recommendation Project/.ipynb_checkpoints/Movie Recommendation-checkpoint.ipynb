## Creating Movie Recommendation

```python
df.head()
```

```python
movie_mat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
movie_mat
```

```python
ratings.sort_values('num of counts', ascending=False).head()
```

```python
starwars_user_ratings = movie_mat['Star Wars (1977)']
```

```python
similar_to_starwars = movie_mat.corrwith(starwars_user_ratings)
```

```python
print(type(similar_to_starwars))
```

```python
similar_to_starwars
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['Correlation'])
```

```python
corr_starwars.dropna()
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars['num of counts'] = ratings['num of counts']
```

```python
corr_starwars
```

```python
corr_starwars[corr_starwars['num of counts']>100].sort_values('Correlation', ascending=False)
```

## Function to predict movies

```python
def predict_movies(movie_name):
    movie_user_ratings = movie_mat[movie_name]
    similar_to_movie = movie_mat.corrwith(movie_user_ratings)
    corr_movie = pd.DataFrame(similar_to_movie, columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie['num of counts'] = ratings['num of counts']
    predictions = corr_movie[corr_movie['num of counts']>100].sort_values('Correlation', ascending=False)
    return predictions    
```

```python
predict_movies('Titanic (1997)').head()
```
