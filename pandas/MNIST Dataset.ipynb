# MNIST Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('mnist_train.csv')
```

```python
data = df.values
```

```python
np.random.shuffle(data)
```

```python
X = data[ : , 0]
Y = data[ : , 1:]
```

```python
print(Y)
```

```python
def imgPlot(X,Y,i):
    plt.imshow(Y[i].reshape((28,28)),cmap='gray')
    plt.title('Label {}'.format(X[i]))
    plt.show
```

```python
imgPlot(X,Y,0)
```

### Splitting of Dataset

```python
split = int(0.80*Y.shape[0])
split
```

```python
X_train,Y_train = X[ :split],Y[ :split, : ]
X_test,Y_test = X[split: ],Y[split: , : ]
```

## Printing the first 25 images in a grid

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(Y[i].reshape(28,28), cmap='gray')
    plt.title(X[i])
    plt.axis('off')
```
